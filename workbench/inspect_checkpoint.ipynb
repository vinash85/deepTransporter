```python
import torch

def inspect_checkpoint(checkpoint_path):
    # Load the checkpoint
    checkpoint = torch.load(checkpoint_path)
    
    # Print available keys in the checkpoint
    print(f"Keys in the checkpoint: {list(checkpoint.keys())}\n")
    
    # Display relevant training information
    if 'train_losses' in checkpoint:
        print(f"Train Losses: {checkpoint['train_losses']}")
    if 'val_losses' in checkpoint:
        print(f"Validation Losses: {checkpoint['val_losses']}")
    if 'train_aucs' in checkpoint:
        print(f"Train AUCs: {checkpoint['train_aucs']}")
    if 'val_aucs' in checkpoint:
        print(f"Validation AUCs: {checkpoint['val_aucs']}")
    if 'best_val_loss' in checkpoint:
        print(f"Best Validation Loss: {checkpoint['best_val_loss']}")
    if 'epoch' in checkpoint:
        print(f"Final Epoch: {checkpoint['epoch']}")
    if 'timestamp' in checkpoint:
        print(f"Saved Timestamp: {checkpoint['timestamp']}")
    if 'hyperparameters' in checkpoint:
        print(f"Hyperparameters: {checkpoint['hyperparameters']}")

# Example usage
checkpoint_path = '/data/servilla/DT_HGNN/Model/Model_Checkpoints/GraphSAGE/2024-09-17_14-32-13_best_model.pt'
inspect_checkpoint(checkpoint_path)
```
