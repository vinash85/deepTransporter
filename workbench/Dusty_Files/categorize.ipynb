```python
# import ace_tools as tools
import pandas as pd
from libchebipy import ChebiEntity

# Load your ChEBI terms from a CSV file (assuming the column name is 'chebi_id')
chebi_df = pd.read_csv(
    '/data_link/servilla/DT_HGNN/data/substrates/ChEBI_184.csv')

# Define categories
categories = {
    'cation': [],
    'anion': [],
    'sugar': [],
    'amino acid': [],
    'protein': [],
    'electron': [],
    'other': []
}

# Function to categorize a term based on its name or definition


def categorize_term(ChEBI_ID):
    try:
        entity = ChebiEntity(ChEBI_ID)
        name = entity.get_name().lower()
        definition = entity.get_definition().lower() if entity.get_definition() else ''

        if 'cation' in name or 'cation' in definition:
            categories['cation'].append(ChEBI_ID)
        elif 'anion' in name or 'anion' in definition:
            categories['anion'].append(ChEBI_ID)
        elif 'sugar' in name or 'sugar' in definition:
            categories['sugar'].append(ChEBI_ID)
        elif 'amino acid' in name or 'amino acid' in definition:
            categories['amino acid'].append(ChEBI_ID)
        elif 'protein' in name or 'protein' in definition:
            categories['protein'].append(ChEBI_ID)
        elif 'electron' in name or 'electron' in definition:
            categories['electron'].append(ChEBI_ID)
        else:
            categories['other'].append(ChEBI_ID)
    except Exception as e:
        categories['other'].append(ChEBI_ID)


# Iterate over each ChEBI term and categorize it
for chebi_id in chebi_df['ChEBI ID']:
    categorize_term(chebi_id)

# Convert categories to DataFrame for easier inspection and saving
categories_df = {category: pd.DataFrame(
    terms, columns=['ChEBI ID']) for category, terms in categories.items()}

# Save each category to a separate CSV file
for category, df in categories_df.items():
    df.to_csv(
        f'/data_link/servilla/DT_HGNN/data/categorized/{category}_chebi_terms.csv', index=False)

# Display the categorized data to the user
# tools.display_dataframe_to_user(
#     name='Categorized ChEBI Terms', dataframe=pd.concat(categories_df))
categories_df

print("ChEBI terms categorized and saved to separate CSV files.")
```

```python
import pandas as pd
from libchebipy import ChebiEntity

# Load your ChEBI terms from a CSV file (assuming the column name is 'ChEBI ID')
chebi_df = pd.read_csv(
    '/data_link/servilla/DT_HGNN/data/substrates/ChEBI_184.csv')

# Define categories
categories = {
    'cation': [],
    'anion': [],
    'sugar': [],
    'amino acid': [],
    'protein': [],
    'electron': [],
    'other': []
}

# Track uncategorized terms
uncategorized = []

# Function to categorize a term based on its name or definition


def categorize_term(chebi_id):
    try:
        entity = ChebiEntity(chebi_id)
        name = entity.get_name().lower()
        definition = entity.get_definition().lower() if entity.get_definition() else ''

        if 'cation' in name or 'cation' in definition:
            categories['cation'].append(chebi_id)
        elif 'anion' in name or 'anion' in definition:
            categories['anion'].append(chebi_id)
        elif 'sugar' in name or 'sugar' in definition:
            categories['sugar'].append(chebi_id)
        elif 'amino acid' in name or 'amino acid' in definition:
            categories['amino acid'].append(chebi_id)
        elif 'protein' in name or 'protein' in definition:
            categories['protein'].append(chebi_id)
        elif 'electron' in name or 'electron' in definition:
            categories['electron'].append(chebi_id)
        else:
            categories['other'].append(chebi_id)
    except Exception as e:
        uncategorized.append(chebi_id)


# Iterate over each ChEBI term and categorize it
for chebi_id in chebi_df['ChEBI ID']:
    categorize_term(chebi_id)

# Convert categories to DataFrame for easier inspection
categories_df = {category: pd.DataFrame(
    terms, columns=['ChEBI ID']) for category, terms in categories.items()}

# Save each category to a separate CSV file
for category, df in categories_df.items():
    df.to_csv(f'{category}_chebi_terms.csv', index=False)

# Check if any terms were left uncategorized
uncategorized_df = pd.DataFrame(uncategorized, columns=['ChEBI ID'])
uncategorized_df.to_csv('uncategorized_chebi_terms.csv', index=False)

print(f"Total entered terms: {len(chebi_df)}")
for category, terms in categories.items():
    print(f"{category}: {len(terms)}")
print(f"Uncategorized terms: {len(uncategorized)}")
```
